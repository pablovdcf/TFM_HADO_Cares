# 1.Codificación de frecuencia

```python
# Codificación de Frecuencia para la columna 'diagnostico'
freq_diagnostico = df['diagnostico'].value_counts(normalize=True)
df['diagnostico'] = df['diagnostico'].map(freq_diagnostico)

# Codificación de Frecuencia para la columna 'motivo_ing'
freq_motivo_ing = df['motivo_ing'].value_counts(normalize=True)
df['motivo_ing'] = df['motivo_ing'].map(freq_motivo_ing)

# Codificación de Frecuencia para la columna 'motivo_alta'
freq_motivo_alta = df['motivo_alta'].value_counts(normalize=True)
df['motivo_alta'] = df['motivo_alta'].map(freq_motivo_alta)

# Codificación de Frecuencia para la columna 's_procedencia'
freq_s_procedencia = df['s_procedencia'].value_counts(normalize=True)
df['s_procedencia'] = df['s_procedencia'].map(freq_s_procedencia)

# Codificación de Frecuencia para la columna 'otros'
freq_otros = df['otros'].value_counts(normalize=True)
df['otros'] = df['otros'].map(freq_otros)

# Codificación de Frecuencia para la columna 'otros_1'
freq_otros_1 = df['otros_1'].value_counts(normalize=True)
df['otros_1'] = df['otros_1'].map(freq_otros_1)

# Codificación de Frecuencia para la columna 'otros_2'
freq_otros_2 = df['otros_2'].value_counts(normalize=True)
df['otros_2'] = df['otros_2'].map(freq_otros_2)

# Codificación de Frecuencia para la columna 'otros_complicaciones'
freq_otros_complicaciones = df['otros_complicaciones'].value_counts(normalize=True)
df['otros_complicaciones'] = df['otros_complicaciones'].map(freq_otros_complicaciones)
```

```python
# Función "check" para ver los valores únicos de cada columna
def unicos(df):
    columns = df.columns.tolist()
    for i, column in enumerate(columns):
        v_column = df[column].unique()
        print(df.columns.values[i])
        print(len(v_column), v_column)
        print("="*66)
unicos(df)
```

```python
columnas = ['ap', 
           'paliativo_onc_noc', 
           'paliativo_no_onc_noc',
           'fiebre',
           'disnea',
           'dolor',
           'delirium',
           'p_terminal',
           'agonia',
           'ast_anorx',
           'agudo_estable',
           'cronico_reag',
           'trato_antibiotico',
           'transfusion',
           'paracentesis',
           'toracocentesis',
           'fe_iv']


def map_values(val):
    if val in ['no', 'vacio', 'b', 'm']:
        return 0
    elif val in ['si', '2 concentrados']:
        return 1
    else:
        return 0
```

```python
for col in columnas:
    df[col] = df[col].apply(map_values)
unicos(df)
```
