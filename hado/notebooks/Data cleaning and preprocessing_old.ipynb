# Quitar espacios

```python
def data_strip(df):
    for column in df:
        df[column] = df[column].replace(r"^ +| +$",r"", regex = True) # Quitamos los espacios del principio y final de los datos
```

# Convertir todo a minúsculas

```python
def to_lowercase(df):
    for column in df:
        if df[column].dtype == 'object':  # comprueba si la columna es de tipo 'object'
            df[column] = df[column].str.lower()
```

```python
to_lowercase(df_values)
```

# Eliminar puntuación y caracteres especiales:

```python
import string

def remove_punctuation(df):
    for column in df:
        if df[column].dtype == 'object':  # comprueba si la columna es de tipo 'object'
            df[column] = df[column].str.translate(str.maketrans("", "", string.punctuation))
```

```python
remove_punctuation(df_values)
```

# Eliminar números (si no son relevantes):

```python
def remove_numbers(df):
    for column in df:
        if df[column].dtype == 'object':  # comprueba si la columna es de tipo 'object'
            df[column] = df[column].str.replace(r'\d+', '', regex=True)
```

# Corregir errores ortográficos (usando la biblioteca TextBlob para corrección automática):

OJO no puede tener valores NA o de otro tipo que no sea object

```python
from textblob import TextBlob

def correct_spelling(df):
    for column in df:
        if df[column].dtype == 'object':  # comprueba si la columna es de tipo 'object'
            df[column] = df[column].apply(lambda x: str(TextBlob(x).correct()))
```
